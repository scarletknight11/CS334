```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
from PIL import Image 
import colorsys
from scipy.cluster.vq import kmeans,vq
from sklearn.cluster import KMeans
from sklearn.metrics import euclidean_distances
from sklearn.datasets import load_sample_image
from sklearn.utils import shuffle
from sklearn.datasets import make_blobs
from time import time
```

```python
flower = 'flowers.tif'
flower_img = cv2.imread(flower)
plt.figure(figsize = [18,5])
plt.imshow(flower_img); plt.title('Original')
```

```python
#RGB values
img_rgb = cv2.cvtColor(flower_img, cv2.COLOR_BGR2RGB)
plt.figure(figsize = [18,5])
plt.imshow(img_rgb); plt.title('RGB Image')
```

```python
#HSV Value
hsvImage = cv2.cvtColor(flower_img, cv2.COLOR_BGR2HSV)
plt.figure(figsize = [18,5])
plt.imshow(hsvImage); plt.title('HSV Image')
```

```python
#CIE xy
CIEXYZ = cv2.cvtColor(flower_img, cv2.COLOR_BGR2XYZ)
plt.figure(figsize = [18,5])
plt.imshow(CIEXYZ); plt.title('CIEXYZ Image')
```

```python
# BGR to CIE L*a*b
cielab = cv2.cvtColor(flower_img, cv2.COLOR_BGR2Lab)
plt.figure(figsize = [18,5])
plt.imshow(cielab); plt.title('CIELAB Image')
```

```python
Fruits = 'Fruits-RGB.tif'
fruits_img = cv2.imread(Fruits)
plt.figure(figsize = [18,5])
plt.imshow(fruits_img); plt.title('Original Fruit Image')
```

```python
img = cv2.imread('Fruits-RGB.tif')
Z = img.reshape((-1,3))
# convert to np.float32
Z = np.float32(Z)
# define criteria, number of clusters(K) and apply kmeans()
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 8
ret,label,center=cv2.kmeans(Z,K,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
# Now convert back into uint8, and make original image
center = np.uint8(center)
res = center[label.flatten()]
res2 = res.reshape((img.shape))
plt.imshow()
```
