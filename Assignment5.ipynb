```python
Fruits = 'Fruits-RGB.tif'
fruits_img = cv2.imread(Fruits)
plt.figure(figsize = [18,5])
plt.imshow(fruits_img); plt.title('Original Fruit Image')
```

```python
img = cv2.imread('Fruits-RGB.tif')
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb); plt.title('RGB Color')
```

```python
def quant(image_rgb, k):
    data = np.float32(image_rgb).reshape(-1,3)
    calculate = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 20, 1.0)
    get, labeldata, compute = cv2.kmeans(data, k, None, calculate, 10, cv2.KMEANS_RANDOM_CENTERS)
    compute = np.uint8(compute)
    result = compute[labeldata.flatten()]
    result = result.reshape(image_rgb.shape)
    return result

plt.imshow(quant(image_rgb,128)); plt.title("K = 128")
```

```python
plt.imshow(quant(image_rgb, 64)); plt.title("K = 64")
```

```python
plt.imshow(quant(image_rgb, 16)); plt.title("K = 16")
```

```python
CIEXYZ2 = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2XYZ)
plt.imshow(quant(CIEXYZ2, 128)); plt.title('CIEXYZ quant 128')
```

```python
CIEXYZ2 = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2XYZ)
plt.imshow(quant(CIEXYZ2, 64)); plt.title('CIEXYZ quant 64')
```
